Obtaining data

```python
import requests
from requests import get
import time
from bs4 import BeautifulSoup
url = 'https://www.smtown.com/album?page='
title_containers = []
name_containers = []
date_containers = []
for page in range(1,87):
    print('---', page, '---')
    r = requests.get(url + str(page))
    response = get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    title = soup.find_all('span', class_ = 'title')
    name = soup.find_all('span', class_ = 'name')
    date = soup.find_all('span', class_ = 'date')
    title_containers += title
    name_containers += name
    date_containers += date
    time.sleep(1)
```

Data Cleaning

```python
for i in range(1028):
    title_containers[i] = title_containers[i].text
    print(title_containers[i])
```

```python
for i in range(1028):
    name_containers[i] = name_containers[i].text
    print(name_containers[i])
```

```python
for i in range(1028):
    #print(date_containers[i].text)
    date_containers[i] = date_containers[i].text
    print(date_containers[i])
```

Write to JSON

```python
import json
album = {}
for i in range(1028):
    album[i+1] = {
        'Title': title_containers[i],
        'Artist': name_containers[i],
        'Date': date_containers[i],
    }
with open('smtown.json', 'w') as outfile: 
    json.dump(album, outfile, indent=4)
```
