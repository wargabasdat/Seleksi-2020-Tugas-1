```python
# import dependencies

import bs4
from urllib.request import urlopen as uReq, Request
from bs4 import BeautifulSoup as soup
import pandas as pd
import time
from tqdm import tqdm
import json
from threading import *
```

```python
# webpage information
base_url_all_hospital = 'https://www.hospitalsafetygrade.org/all-hospitals'

# http header
header = {'user-agent' : 'Mozilla/5.0 (X11; Linux x86_64); Kevin/18218001@std.stei.itb.ac.id'}
```

```python
# array to store information
array_link_rs = []
array_name_rs = []

array_address_rs1 = []
array_grade_rs1 = []
array_score_infection1 = []
array_score_problemsWithSurgery1 = []
array_score_practicesToPreventErrors1 = []
array_score_safetyProblems1 = []
array_score_doctorNurseStaff1 = []

array_address_rs2 = []
array_grade_rs2 = []
array_score_infection2 = []
array_score_problemsWithSurgery2 = []
array_score_practicesToPreventErrors2 = []
array_score_safetyProblems2 = []
array_score_doctorNurseStaff2 = []

array_address_rs3 = []
array_grade_rs3 = []
array_score_infection3 = []
array_score_problemsWithSurgery3 = []
array_score_practicesToPreventErrors3 = []
array_score_safetyProblems3 = []
array_score_doctorNurseStaff3 = []

array_address_rs4 = []
array_grade_rs4 = []
array_score_infection4 = []
array_score_problemsWithSurgery4 = []
array_score_practicesToPreventErrors4 = []
array_score_safetyProblems4 = []
array_score_doctorNurseStaff4 = []
```

## Get Hospital Data

```python
# connect and download html base url webpage
req = Request(base_url_all_hospital, headers = header)
uClient = uReq(req)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
```

```python
# find element contain the name and hyperlink of the hospital
hospitalNameAndLink = page_soup.findAll("div", {"id": "BlinkDBContent_849210"})[0].findAll("a")

# put the name into array_name_rs and the hyperlink to array_link_rs
for hospital in hospitalNameAndLink :
    hospitalName = hospital.text
    hospitalLink = hospital['href']
    array_name_rs.append(hospitalName)
    array_link_rs.append(hospitalLink)
```

```python
# define a function to begin scraping
def scrap(array_link_rs_processed, array_grade_rs, array_address_rs, array_score_infection, array_score_problemsWithSurgery, array_score_practicesToPreventErrors, array_score_safetyProblems, array_score_doctorNurseStaff) :
    
    header = {'user-agent' : 'Mozilla/5.0 (X11; Linux x86_64); Kevin/18218001@std.stei.itb.ac.id'}
    # get information (address, grades, score for each infections)
    for link in tqdm(array_link_rs_processed) :
        # get html
        temp_url = link
        req = Request(temp_url, headers = header)
        uClient = uReq(req)
        page_html = uClient.read()
        uClient.close()
        page_soup = soup(page_html, "html.parser")

        # find grades
        grades = [grade.img["alt"].split(' ')[1] for grade in page_soup.findAll("div", {"class": "grade"})]
        array_grade_rs.append(grades)

        # find address
        addressHTML = page_soup.findAll("div", {"class": "address"})[0]
        address = ' '.join(''.join(addressHTML.text.split('\n')[:-2]).split()[:-1])
        array_address_rs.append(address)

        # find all score
        majorCriteria = ["c1","c2","c3","c4","c5"]
        temp_parent_array = []
        for eachCriteria in majorCriteria :
            temp_array = []
            scoreHTML = page_soup.findAll("div", {"data-tab-content-id": eachCriteria})
            for criteria in scoreHTML :
                score = criteria.findAll("p", {"class": "value"})[0].text
                if (score == 'Not Available') :
                    score = -1.0
                elif (score == 'Declined to Report') :
                    score = -2.0
                else :
                    score = float(score)
                temp_array.append(score)
            temp_parent_array.append(temp_array)
        if (len(temp_parent_array) != 5) :
            temp_parent_array = [[],[],[],[],[]]
        array_score_infection.append(temp_parent_array[0])
        array_score_problemsWithSurgery.append(temp_parent_array[1])
        array_score_practicesToPreventErrors.append(temp_parent_array[2])
        array_score_safetyProblems.append(temp_parent_array[3])
        array_score_doctorNurseStaff.append(temp_parent_array[4])
        time.sleep(1)
```

```python
# define thread classes and split all hospital into 4 classes
class Quarter1(Thread) :
    def run(self) :
        scrap(array_link_rs[:656], array_grade_rs1, array_address_rs1, array_score_infection1, array_score_problemsWithSurgery1, array_score_practicesToPreventErrors1, array_score_safetyProblems1, array_score_doctorNurseStaff1)
        
class Quarter2(Thread) :
    def run(self) :
        scrap(array_link_rs[656:1311], array_grade_rs2, array_address_rs2, array_score_infection2, array_score_problemsWithSurgery2, array_score_practicesToPreventErrors2, array_score_safetyProblems2, array_score_doctorNurseStaff2)
        
class Quarter3(Thread) :
    def run(self) :
        scrap(array_link_rs[1311:1966], array_grade_rs3, array_address_rs3, array_score_infection3, array_score_problemsWithSurgery3, array_score_practicesToPreventErrors3, array_score_safetyProblems3, array_score_doctorNurseStaff3)
        
class Quarter4(Thread) :
    def run(self) :
        scrap(array_link_rs[1966:], array_grade_rs4, array_address_rs4, array_score_infection4, array_score_problemsWithSurgery4, array_score_practicesToPreventErrors4, array_score_safetyProblems4, array_score_doctorNurseStaff4)
```

```python
# begin scraping in all threads

t1 = Quarter1()
t2 = Quarter2()
t3 = Quarter3()
t4 = Quarter4()

t1.start()
t2.start()
t3.start()
t4.start()

t1.join()
t2.join()
t3.join()
t4.join()

print("Scraping Done")
```

```python
# merge all subcomponents into one component

array_address_rs = array_address_rs1 + array_address_rs2 + array_address_rs3 + array_address_rs4
array_grade_rs = array_grade_rs1 + array_grade_rs2 + array_grade_rs3 + array_grade_rs4
array_score_infection = array_score_infection1 + array_score_infection2 + array_score_infection3 + array_score_infection4
array_score_problemsWithSurgery = array_score_problemsWithSurgery1 + array_score_problemsWithSurgery2 + array_score_problemsWithSurgery3 + array_score_problemsWithSurgery4
array_score_practicesToPreventErrors = array_score_practicesToPreventErrors1 + array_score_practicesToPreventErrors2 + array_score_practicesToPreventErrors3 + array_score_practicesToPreventErrors4
array_score_safetyProblems = array_score_safetyProblems1 + array_score_safetyProblems2 + array_score_safetyProblems3 + array_score_safetyProblems4
array_score_doctorNurseStaff = array_score_doctorNurseStaff1 + array_score_doctorNurseStaff2 + array_score_doctorNurseStaff3 + array_score_doctorNurseStaff4
```
